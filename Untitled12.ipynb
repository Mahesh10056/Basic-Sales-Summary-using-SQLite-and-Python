```python
import sqlite3

conn = sqlite3.connect('sales_data.db')
cursor = conn.cursor()

# Create table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS sales (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        product TEXT,
        quantity INTEGER,
        price REAL
    )
''')

# Insert sample data
data = [
    ('Pen', 10, 5.0),
    ('Notebook', 5, 20.0),
    ('Pencil', 15, 2.0),
    ('Eraser', 8, 3.0),
    ('Notebook', 7, 20.0),
    ('Pen', 12, 5.0)
]
cursor.executemany('INSERT INTO sales (product, quantity, price) VALUES (?, ?, ?)', data)
conn.commit()
conn.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

# Connect to the DB
conn = sqlite3.connect('sales_data.db')

# SQL query
query = '''
SELECT product,
       SUM(quantity) AS total_qty,
       SUM(quantity * price) AS revenue
FROM sales
GROUP BY product
'''

# Load result into a DataFrame
df = pd.read_sql_query(query, conn)
conn.close()

# Print result
print(df)

# Plot
df.plot(kind='bar', x='product', y='revenue', color='lightgreen', legend=False)
plt.title('Revenue by Product')
plt.ylabel('Revenue')
plt.tight_layout()
plt.savefig('sales_chart.png')  # Saves the chart image
plt.show()
```
